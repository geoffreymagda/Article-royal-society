```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import pandas as pd
import matplotlib.patches as mpatches
from matplotlib.colors import LightSource
%matplotlib notebook
mu0=4*np.pi/10000000

chiv = 2500
mu = mu0*(1 + chiv)
Jsat = 2.15
ms = Jsat/mu0
chir = chiv/(1 + chiv)
beta = 3*chir/mu0/ms

def m(b):
    return ms*np.tanh(b*chir/mu0/ms)
```

```python
blist=np.arange(0,6,0.1)
plt.figure(figsize=(5,4))
plt.plot((blist/mu0 - m(blist))/ms*chiv,m(blist)/ms,label="m")
plt.ylabel("||m||/$m_{s}$")
plt.xlabel("||h||/$h_{ref}$")
plt.plot((blist/mu0 - m(blist))/ms*chiv,np.ones(len(blist)),label="$m_{s}$",color='k',linestyle='dashed')
plt.legend(loc="right")
plt.savefig('/home/magda/Documents/GitHub/thesis-redux/non linear wire b  para/results final law larger 2/figs/magnetisation far.pdf') 
```
